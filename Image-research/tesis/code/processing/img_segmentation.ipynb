```python
from sklearn.cluster import KMeans
def kMeans_clustering(img, n_clusters):
    # Reshape the image to a 2D array of pixels
    img_flat = img.reshape(-1, 3)
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, n_init=10)
    kmeans.fit(img_flat)
    # Map the clusters back to image space
    segmented_img = kmeans.cluster_centers_[kmeans.labels_].reshape(img.shape).astype(np.uint8)
    return segmented_img
```

```python
data_dir = "D:/pythonProject/datasets/test-dataset/test"
for img_file in os.listdir(data_dir):
    print("--------------------------------------------------------")
    img_path = os.path.join(data_dir, img_file)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    segmented_img = kMeans_clustering(img_rgb, 3)
            
    display_imgs([img_rgb, segmented_img], ["Original Image", "K-means Segmented Image"])
```
