```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2

from preprocessing import rgb_to_hsv, rgb_to_gray
from shared import display_imgs, display_plots
```

```python
def histogram_equalization(img):
    eq = cv2.equalizeHist(img)
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])
    eq_hist = cv2.calcHist([eq], [0], None, [256], [0, 256])
    
    display_plots([hist, eq_hist], ["hist", "hist equalized"])

    return eq
```

```python
data_dir = "D:/pythonProject/datasets/test-dataset/test"
for img_file in os.listdir(data_dir):
    print("--------------------------------------------------------")
    img_path = os.path.join(data_dir, img_file)
    img = cv2.imread(img_path)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    h, s, v = cv2.split(img_hsv)
    eq = histogram_equalization(v)
    img_hist = cv2.merge([h, s, eq])
    
    print("COLOR_BGR2HSV")
    display_imgs([h, s, v], ["h", "s", "v"])
    display_imgs([img, img_hsv, img_hist], ["Original Image", "COLOR_BGR2HSV", "histogram_equalization"])
    
    print("--------------------------------------------------------")
```
